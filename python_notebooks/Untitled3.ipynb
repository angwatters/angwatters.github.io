```python
import pandas as pd
import bqplot
import numpy as np
import ipywidgets
import matplotlib.pyplot as plt
import geopandas
```

```python
ufo = pd.read_csv("modified_ufo.csv")

ufo.head()
```

```python
ufo['latitude'] = pd.to_numeric(ufo['latitude'],errors='coerce')
ufo['longitude'] = pd.to_numeric(ufo['longitude'],errors='coerce')

gdf = geopandas.GeoDataFrame(ufo, geometry=geopandas.points_from_xy(ufo.longitude, ufo.latitude))
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
```

```python
fig, ax = plt.subplots(figsize=(20,10))

world.plot(ax=ax, edgecolor='k', color='darkblue')
gdf.plot(ax=ax, color='lightgreen')

plt.show()
```
